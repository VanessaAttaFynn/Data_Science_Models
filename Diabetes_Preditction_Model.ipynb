## Handling Missing Values

```python
attr = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for i in attr:
    df[i].replace(0, df[i].mean(), inplace=True)
```

```python
df.head(20)
```

```python
#Missing values have now been replace my the mean of each column
```

## Detecting Outliers

```python
df.plot(kind='box', figsize=(12,12), layout=(3,3), subplots=True);
```

```python
#Handling Outliers
df.describe()
```

```python
df['Glucose'] = np.where(df['Glucose']>=140.250, 140.250, df['Glucose'])
df['BloodPressure'] = np.where(df['BloodPressure']>=80, 80, df['BloodPressure'])
df['SkinThickness'] = np.where(df['SkinThickness']>=32, 32, df['SkinThickness'])
df['Insulin'] = np.where(df['Insulin']>=127.250, 127.250, df['Insulin'])
df['BMI'] = np.where(df['BMI']>=36.60, 36.60, df['BMI'])
df['DiabetesPedigreeFunction'] = np.where(df['DiabetesPedigreeFunction']>=0.626250, 0.626250, df['DiabetesPedigreeFunction'])
```

```python
df.plot(kind='box', figsize=(12,12), layout=(3,3), subplots=True);
```

```python
df.describe()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler().fit(df.drop(["Outcome"], axis=1))

x = ss.transform(df.drop(["Outcome"], axis=1))
y = df["Outcome"]
```

```python
# Correlation
df.corr()
```

```python
# Heatmap
sns.heatmap(df.corr(), annot= True)
```

```python
# From the heatmap we see that the corrention between the independent var and outcome is weak.
# The attribute with the strongest relationship with outcome is Age
```

## Model Building

```python
# The dataset must now be split into test and training set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=4)
```

### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#Measuring Accurancy
lr = LogisticRegression()
model = lr.fit(x_train, y_train)
prediction = model.predict(x_test)

print("Acc. on trainig data: {:,.3f}".format(lr.score(x_train, y_train)))
print("Acc. on test data: {:,.3f}".format(lr.score(x_test, y_test)))
```

### 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion='entropy',max_depth=4)
model1 = dtc.fit(x_train, y_train)
prediction1 = model1.predict(x_test)

print("Acc. on training data: {:,.3f}".format(dtc.score(x_train, y_train)))
print("Acc. on test data: {:,.3f}".format(dtc.score(x_test, y_test)))
```

### 3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion='entropy',max_depth=4)
model2 = rfc.fit(x_train, y_train)
prediction2 = model2.predict(x_test)
print("Acc. on training data: {:,.3f}".format(rfc.score(x_train, y_train)))
print("Acc. on test data: {:,.3f}".format(rfc.score(x_test, y_test)))
```
